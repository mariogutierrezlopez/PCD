```python
import threading, os, platform, requests, json
```

```python
url1 = "https://github.com/sukuzhanay"
url2 = "https://www.linkedin.com/in/sucuzhanay/"
url3 = "https://www.credly.com/users/christian-vladimir-sucuzhanay-arevalo/badges"
url4 = "https://stackoverflow.com/users/6025932/christian-vladimir-sucuzhanayLinks"
url5 = "https://www.cloudskillsboost.google/public_profiles/c094dd66-f266-4f77-a808-a0601b187009"

hilos1 = 100
hilos2 = 50
hilos3 = 20
hilos4 = 30
hilos5 = 20
```

```python
#Pagina 1
def matarpag1():
    print(requests.get(url1))
    threading.current_thread()

for i in range(hilos1):
    threading.Thread(target=matarpag1()).start()

#Pagina 2
def matarpag2():
    print(requests.get(url2))
    threading.current_thread()

for i in range(hilos2):
    threading.Thread(target=matarpag2()).start()

#Pagina 3
def matarpag3():
    print(requests.get(url3))
    threading.current_thread()

for i in range(hilos3):
    threading.Thread(target=matarpag1()).start()

#Pagina 4
def matarpag4():
    print(requests.get(url4))
    threading.current_thread()

for i in range(hilos4):
    threading.Thread(target=matarpag4()).start()
    
#Pagina 5
def matarpag5():
    print(requests.get(url5))
    threading.current_thread()

for i in range(hilos5):
    threading.Thread(target=matarpag5()).start()
```
